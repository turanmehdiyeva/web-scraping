```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
lstn = []
lstp = []
lstd = []
d4 = {}
for j in range(1,5):
    url = 'https://irshad.az/az/telefon-ve-plansetler/mobil-telefonlar/?from=300&to=1000&sort=1&p={}'.format(j)
    page = requests.get(url)
    
    soup = BeautifulSoup(page.content, 'html.parser')
    info = soup.find('div', class_ = 'products_grid')
    info_rows = info.find_all('figure', class_='pr_self')

    for index, row in enumerate(info_rows):
        lstn.append(row.find('b').get_text().replace('Mobil telefon','').strip())
        d4['phones'] = lstn
        lstp.append(row.find('div', class_="two_price").get_text(' ', strip=True).replace('\n',''))
        d4['Price'] = lstp
        try:
            lstd.append(row.find('div', class_='discount_under_price').get_text(' ', strip=True))
        except:
            lstd.append('None')
        d4['Discount'] = lstd
```

```python
pd.DataFrame(d4)
```
